```
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,6,8]
y=[3,8,1,10]
plt.plot(x,y)
plt.title("Graph")
# plt.xlabel('x-axis')
# plt.ylabel('y-axis')
plt.grid()
plt.show()
```

```
x1=[1,2,3,4,5]
x2=[4,6,1,8,3]
y1=[2,5,4,6,7]
y2=[7,1,2,3,4]
plt.plot(x1,x2,y1,y2)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,6,8]
y=[3,8,1,10]
plt.plot(x,y,linestyle='dotted',linewidth=20)
plt.title("Graph")

plt.grid()
plt.show()
```

```
plt.bar(x,y,width=0.1)
```

```
plt.barh(x,y,color='green',height=0.8)
```

```
plt.plot(x,y,'o-.r',marker='o',ms=10,mec='b',mfc='m') #-,:,--,-.  markersize, markeredgecolor, markerfacecolor
```

```

labels=['sunflower','lily','rose','dandelion']
plt.pie(x,labels=labels)
plt.legend()
plt.show()
```

```
explode=[0.3,0.1,0.1,0.2]
plt.pie(x,labels=labels,explode=explode,shadow=True)
plt.show()
```
